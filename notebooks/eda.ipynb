```python
sns.countplot(x="Class", data=df)
plt.title("Class Distribution")
plt.show()
```

```python
plt.figure(figsize=(8,4))
sns.histplot(df["Amount"], bins=50)
plt.title("Amount Distribution")
plt.show()
```

```python
plt.figure(figsize=(8,4))
sns.histplot(np.log1p(df["Amount"]), bins=50)
plt.title("Log-transformed Amount")
plt.show()
```

Better to use log Amount!

```python
plt.figure(figsize=(8,4))
sns.histplot(np.log1p(df[df["Class"] == 0]["Amount"]), color="blue", label="Normal", bins=50, alpha=0.6)
sns.histplot(np.log1p(df[df["Class"] == 1]["Amount"]), color="red", label="Fraud", bins=50, alpha=0.6)
plt.legend()
plt.title("Amount: Fraud vs Normal")
plt.show()
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(df.corr(), cmap="coolwarm")
plt.title("Correlation Matrix")
plt.show()
```

Correlation is very weak among features, since they are already decorrelated by PCA. However, there some correlation between some PCA features with Amount, Class, Time, since these three features are not PCA.
